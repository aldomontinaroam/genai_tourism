```python
import pandas as pd
import re
import random
import numpy as np

# Sample dataframe
df = pd.read_csv('/Users/aldo.montinaro/Desktop/sci_stuff/ai_planning_by_european_countries.csv')


df.head()
```

```python
# Drop rows where Start date is NaN
df = df.dropna(subset=['Start date'])

# Convert Start date column to integers
df['Start date'] = df['Start date'].astype(int)
```

```python
# Function to convert budget range to integer
def convert_budget(budget_range):
    if budget_range == "Less than 1M":
        return random.randint(0, 999999)
    elif budget_range == "More than 500M":
        return 500000000
    else:
        lower, upper = budget_range.split('-')
        return random.randint(int(lower[:-1]) * 1000000, int(upper[:-1]) * 1000000)

# Apply function to Yearly budget range column
df['Yearly budget'] = df['Yearly budget range'].apply(convert_budget)

# Drop Yearly budget range column
df.drop(columns=['Yearly budget range'], inplace=True)

df
```

```python
df.to_csv('policy_budget.csv')
```
